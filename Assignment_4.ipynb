## Question 1

Explain the difference between Encoder-only and Decoder-only architectures with a real-world example not discussed in class.

**Answer:**

Encoder-only architectures process an entire input sequence simultaneously to understand its rich context (bidirectional).

For eg: In a Google search, when we type a search query, the search engine needs to understand the exact context.

Decoder-only architectures generate text one token at a time by predicting the next word based only on preceding words.

For eg: When we ask ChatGPT to write a reply mail to an HR, the model needs to generate new text from scratch. Based on the prompt which we provide it, it predicts the next most likely word and goes on.

## Question 2

Why is bidirectional context useful for text understanding? Give two NLP tasks where it is beneficial.

**Answer:**

Bidirectional context is crucial for text understanding because many words are ambiguous or dependent on subsequent text, meaning their true intent can only be determined by looking at both the words before and after them. In language, tokens don't just inherit meaning from the past; the future tokens actively shape them. If a model only reads from left to right (unidirectional), it loses half the context required to resolve ambiguities, homonyms, or complex grammatical dependencies.

Two NLP tasks where it is beneficial:

1. **Named Entity Recognition (NER)**

NER involves identifying and classifying proper nouns into categories like people, organizations, or locations.

Use of bidirectionality:

Eg:
(i) "Apple announced their new mixed-reality headset today."

(ii) "Apple pie is my favorite dessert."

A left-to-right model reading the second sentence sees "Apple" at the start and might immediately guess it's the tech company. A bidirectional model looks ahead to the word "pie" and instantly corrects its understanding, classifying it as a fruit/food item rather than an organization.

2. **Coreference Resolution**

This task involves determining which nouns or pronouns refer to the same entity in a text.

Use of bidirectionality:

Eg:
(i) "The trophy didn't fit into the brown suitcase because it was too large."

(ii)	"The trophy didn't fit into the brown suitcase because it was too small."

To understand what "it" refers to, the model cannot just look at the preceding words ("trophy" and "suitcase" are present in both). It must look ahead to the adjectives at the very end—"large" or "small". By processing the sentence bidirectionally, the model easily deduces that in the first sentence, "it" is the trophy (too large to fit), and in the second sentence, "it" is the suitcase (too small to house it).

## Question 3

Why can't a Decoder see future tokens during training? Explain with an example.

**Answer:**

A Decoder cannot see future tokens during training because doing so would allow it to "cheat" by looking up the correct answer instead of learning how to predict it, rendering the model useless during real-world inference when future words don't exist yet. This restriction is enforced to use this model to actually predict tokens.

Consider the following sentence that a decoder model is being trained to predict:

"The cat sat on the mat."

During training, the model is fed the tokens sequentially, and its goal at each step is to predict the next token,

If the model is trying to predict the word after 'cat', and can see the future tokens, the attention mechanism will naturally locate the token immediately following 'cat', which is 'sat'.

In this way, instead of actually learning the patterns of a language, grammer or logic, the model simply copies the next word from the input sentence and achieves a 100% training accuracy while actually learning nothing and hence becoming useless for actual use.

## Question 4

Compare BERT and GPT in terms of architecture, training objective, input visibility, and applications.

**Answer:**

BERT is an encoder-only model designed to deeply understand text by looking at it bidirectionally, while GPT is a decoder-only model built to generate text one word at a time by looking only at the past.

1. Architecture

•	BERT: Built entirely out of Transformer encoder blocks. It doesn’t generate new text; it just creates vector representations of the text you feed it.

•	GPT: Built out of Transformer decoder blocks. It takes its own previous output and feeds it back into the input to keep writing.

2. Training Objective

•	BERT: Trained on "Masked Language Modeling" (MLM). It is like a fill-in-the-blank test. It hides random words in a sentence (e.g., "The [MASK] sat on the mat") and tries to guess them.

•	GPT: Trained on standard "Causal Language Modeling." It's basically a next-word predictor. Given "The cat sat on the", it has to guess "mat".

3. Input Visibility

•	BERT: It has full 360-degree vision. When processing a word, it can read both the words before it and the words after it (bidirectional).

•	GPT: It can only look at words to its left. It is physically blocked from seeing any "future" tokens during training so it doesn't cheat.

4. Applications

•	BERT: It is used for Google Search query understanding, sentiment analysis, extracting specific names/dates from documents, or classification.

•	GPT: It is used for writing essays, drafting emails, writing code, or building interactive chatbots.

## Question 7

What is Masked Self-Attention? Explain why it prevents cheating.

**Answer:**

* **Masked Self-Attention** is a mechanism used in decoder-only models (such as GPT) that allows the model to attend only to the current and previous tokens while blocking access to future tokens.

* It prevents cheating by hiding future words during training, ensuring the model predicts the next word using only the information it has already seen.

**Example:**
For the sentence **"I love playing football."**, when predicting **"playing"**, the model can only see **"I love"** and cannot see **"football."** This forces it to learn genuine next-word prediction instead of relying on future information.

## Question 8

Differentiate Parameters, Weights, and Biases with examples.

**Answer:**

* **Parameters:** Parameters are the **learnable values** in a neural network that are updated during training. They include both **weights and biases**.

  * **Example:** If a model has 1 million weights and 1,000 biases, it has **1,001,000 parameters**.

* **Weights:** Weights determine the **importance or influence** of each input feature on the output. They are multiplied with the input values.

  * **Example:** In predicting house prices, the weight for **house size** may be higher than the weight for **number of windows**, indicating that size has a greater impact.

* **Biases:** Biases are additional values added to the weighted sum, allowing the model to make predictions even when all input values are zero.

  * **Example:** In the equation **Output = (Weight × Input) + Bias**, if **Weight = 2**, **Input = 5**, and **Bias = 3**, then **Output = (2 × 5) + 3 = 13**.

## Question 9

If a model has 15 billion parameters stored in FP16, estimate the memory required.

**Answer:**

**Given:**

* Number of parameters = **15 billion = 15 × 10⁹**
* Data type = **FP16 (16-bit floating point)**
* **FP16 = 16 bits = 2 bytes per parameter**

**Calculation:**

Memory required = Number of parameters × Bytes per parameter

= **15 × 10⁹ × 2 bytes**

= **30 × 10⁹ bytes**

≈ **30 GB** (or about **27.9 GiB** using binary units)

**Final Answer:**
A model with **15 billion parameters** stored in **FP16** requires approximately **30 GB of memory** (about **27.9 GiB**).

## Question 17

Interview Q1: Why do larger parameter counts not always produce better performance?

**Answer:**

Larger parameter counts do not always produce better performance because:

* **Data Quality:** A large model trained on poor-quality or insufficient data may perform worse than a smaller model trained on high-quality data.
* **Overfitting:** More parameters can cause the model to memorize training data instead of learning general patterns, reducing performance on new data.
* **Higher Computational Cost:** Large models require more memory, processing power, and training time, making them less efficient.
* **Diminishing Returns:** After a certain size, increasing parameters often results in only small improvements in accuracy.
* **Task Suitability:** Some simple tasks do not require very large models; a smaller model can achieve similar or even better performance more efficiently.

## Question 18

Interview Q2: What happens if model weights are randomly initialized and never trained?

**Answer:**

If model weights are randomly initialized and never trained:

* The model will produce **random and unreliable predictions** because it has not learned any patterns from the training data.
* It will **not understand the relationships** between inputs and outputs.
* The model's accuracy will be **very low**, often close to random guessing.
* It will fail to perform tasks such as **classification, translation, or text generation** correctly.
* Training is necessary to update the weights so the model can learn meaningful patterns and make accurate predictions.

## Question 20

Mini Project: Build a notebook demonstrating BERT fill-mask and GPT text generation, compare outputs, limitations, runtime, and conclusions.

**Answer:**

````python
from transformers import pipeline
import time
````

````python
# Load BERT model
fill_mask = pipeline("fill-mask", model="bert-base-uncased")

sentence = "The students are studying for the [MASK]."

start = time.time()
result = fill_mask(sentence)
end = time.time()

print("BERT Output:")
for r in result:
    print(r["token_str"], "-", round(r["score"],4))

bert_time = end - start
````

````python
#Load GPT2 Model
generator = pipeline("text-generation", model="gpt2")

prompt = "Education is important because"

start = time.time()
result = generator(prompt, max_length=30)
end = time.time()

print("\nGPT Output:")
print(result[0]["generated_text"])

gpt_time = end - start
````

````python
print("BERT Runtime:", round(bert_time,3), "seconds")
print("GPT Runtime :", round(gpt_time,3), "seconds")
````
